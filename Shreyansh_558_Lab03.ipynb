```python
hindi = '''यहाँ हिंदी में एनएलपी के लिए कुछ पाठ दिया जा रहा है:

"प्रारंभिक एनएलपी (नेचुरल लैंग्वेज प्रोसेसिंग) एक उपकरण है जो मानवीय भाषा को समझने और उससे संबंधित संदेशों को विश्लेषित करने के लिए डिजाइन किया गया है। यह एक कम्प्यूटरी प्रोसेसिंग उपकरण है जो लिखित या बोली गई मानवीय भाषा को समझ सकता है और उस पर काम कर सकता है। एनएलपी के अनुप्रयोग क्षेत्र में शामिल हैं अनुवाद, संधि-संधान, समाज विज्ञान, शिक्षा, विज्ञान, आर्थिक, विपणन, गहन शोध और भाषा संबंधित बहुत कुछ।"

आशा है कि यह आपके एनएलपी परियोजनाओं के लिए उपयोगी होगा।'''
```

```python
!pip install cltk
```

```python
from cltk import NLP
cltk_nlp = NLP(language="hin")
cltk_doc = cltk_nlp.analyze(text=hindi)
```

```python
import pandas as pd

df = pd.DataFrame(columns=["INDEX", "TEXT", "TOKENS", "LEMMA", "STOP"])
df.TOKENS = cltk_doc.tokens

for i in range(len(cltk_doc.words)):
    df.INDEX[i] = cltk_doc.words[i].index_token
    df.TEXT[i] = cltk_doc.words[i].string
    df.LEMMA[i] = cltk_doc.words[i].lemma
    df.STOP[i] = cltk_doc.words[i].stop
```

```python
Russian = '''Привет! Как дела? Язык – это инструмент, с помощью которого мы выражаем свои мысли и чувства. Нейролингвистическое программирование (NLP) помогает понять, как мы используем язык для взаимодействия с окружающим миром и друг с другом. Исследование текста на русском языке может дать ценные знания о его структуре, семантике и контексте. Надеюсь, этот текст будет полезен для вашего проекта по анализу естественного языка!'''
```

```python
!python -m spacy download ru_core_news_sm
```

```python
import spacy
nlp = spacy.load("ru_core_news_sm")
doc = nlp(Russian)
df = pd.DataFrame(columns=["TEXT", "LEMMA", "POS", "TAG", "DEP", "SHAPE", "ALPHA", "STOP"])
for token in doc:
     df = df.append({
        "TEXT": token.text,
        "LEMMA": token.lemma_,
        "POS": token.pos_,
        "TAG": token.tag_,
        "DEP": token.dep_,
        "SHAPE": token.shape_,
        "ALPHA": token.is_alpha,
        "STOP": token.is_stop
    }, ignore_index=True)
```
