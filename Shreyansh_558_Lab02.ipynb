#i. Find all dates and convert them to the DD-MM-YYYY format.

```python
#find date
pattern = r'\b(\d{1,2})-(january|february|march|april|may|june|july|august|september|october|november|december)-(\d{4})\b'
#change the format
def newDate(match):
    months = {
        "january": "01",
        "february": "02",
        "march": "03",
        "april": "04",
        "may": "05",
        "june": "06",
        "july": "07",
        "august": "08",
        "september": "09",
        "october": "10",
        "november": "11",
        "december": "12"
    }
    day = match.group(1)
    month = months[match.group(2).lower()]
    year = match.group(3)
    return f"{day}-{month}-{year}"
t3 = re.sub(pattern, newDate, t2, flags=re.IGNORECASE)
print(t3)
```
